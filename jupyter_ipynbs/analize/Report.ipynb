```python
plants = [
    "A.thaliana",
    "C.sinensis",
    "G.max",
    "G.raimondii",    
    "M.truncatula",
    "O.sativa",
    "S.bicolor",
    "T.aestivum",
    "Z.mays"    
]
```

```python
def show(plant, _type):    
    temp_path = f"./Experiment/{plant}/{_type}/Temp"
    result_path = f"./Experiment/{plant}/{_type}/Result"
    ext = fasta_to_df(f'./{temp_path}/extended_modified.txt')
    print(f'total:                  {ext.shape[0]}')    
    non_coding = fasta_to_df(f'./{temp_path}/extended_modified_non_coding.txt')
    print(f'non_coding:             {non_coding.shape[0]}')
    coding = fasta_to_df(f'./{temp_path}/extended_modified_coding.txt')
    print(f'coding:                 {coding.shape[0]}')
            
    base = f"{result_path}/secondary_structure/mfold/"
    if(_type == "positive"):
        df = fasta_to_df(f'./{temp_path}/extended_modified_non_coding.txt')
    else:        
        df = fasta_to_df(f'./{temp_path}/extended_modified_coding.txt')
    count_list = []
    for index, row in df.iterrows():    
        tag = reformat(row['tag'])         
        count_list.append(len(glob.glob(f'{base + tag}/*.ct')))        
    print(f"number of structures:   {pd.Series(count_list).sum()}")
    df = pd.read_csv(f"./{result_path}/result_level1_filter.csv")
    print(f'number of filter 1:     {df.shape[0]}')
```

```python
for plant in plants:    
    line = "##############################################"
    print(line)    
    print("".join([" " for i in range((len(line) - len(plant) )// 2)]) + plant)        
    for _type in ["positive", "negative"]:
        print()
        print("".join([" " for i in range((len(line) - len(_type) )// 2)]) + _type)                
        show(plant, _type)
    print()
```
