```python
import numpy as np
import pandas as pd
import os, sys
import seaborn as sns
import matplotlib.pyplot as plt
sys.path.append("./src/")
from convertor import convert
from preprocessing import get_target, preprocessing
from numpy.linalg import norm
```

```python
plant_list = [
    "A.thaliana",
    "C.sinensis",
    "G.max",
    "G.raimondii",
    "M.truncatula",
    "O.sativa",
    "S.bicolor",
    "T.aestivum",
    "Z.mays"
]
```

```python
b = {}
for plant in plant_list:
    data[plant] = {}
    # positive
    df = pd.read_csv(f'./Experiment/{plant}/positive/Result/result_level1_filter.csv')
    df = convert(df)    
    df = df[df['hit len'] < 25]    
    [feature, mu, std]  = preprocessing(df, None, None, False)
    data[plant]["positive"] = feature
    
    # negative
    df = pd.read_csv(f'./Experiment/{plant}/negative/Result/result_level1_filter.csv')
    df = convert(df)    
    df = df[df['hit len'] < 25]
    [feature, mu, std]  = preprocessing(df, None, None, False)
    data[plant]["negative"] = feature
    
    print(f'{plant} finished.')
```

```python
# convert to dataframe
rows = []
for plant in plant_list:
    pos = data[plant]['positive']
    pos['species'] = plant
    pos['microRNA'] = "Yes"
    rows.append(pos)
    neg = data[plant]['negative']
    neg['species'] = plant
    neg['microRNA'] = "No"
    rows.append(neg)    
data = pd.concat(rows)
```
